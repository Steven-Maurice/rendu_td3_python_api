```python
# Exercise 3.1
# Parsing the response from the nationalize API
# Get the most probable country and its percentage
# Example response: {"name":"YOUR_NAME","country":[{"country_id":"FR","probability":0.75},{"country_id":"BE","probability":0.25}]}
# Hint: use the max() function with a lambda function
# https://docs.python.org/3/library/functions.html#max
# https://docs.python.org/3/tutorial/controlflow.html#lambda-expressions

#######################################################
first_name = input("Enter your first name: ")

url = f"https://api.nationalize.io/?name={first_name}"
response = requests.get(url)

if response.status_code == 200:
    response_json = response.json()

    print(f"Answer status_code: {response.status_code}")

    print(type(response_json))

    print(response_json)

    # Parsing the response to get the most probable country and its percentage
    countries = response_json["country"]
    most_probable_country = max(countries, key=lambda x: x["probability"])
    country_id = most_probable_country["country_id"]
    probability = most_probable_country["probability"]

    # Print the most probable country and its percentage
    print(f"The most probable country for {first_name} is {country_id} with a probability of {probability}.")
else:
    print("Error occurred while retrieving data from the API.")
#######################################################
```

```python
# Exercise 3
# Now that we have the ids for all the places in Derbyshire, we can load all their details...
# And from their details, we can list all the details of their manors.
# Go fetch the data!
# P.S.: remember to save the data to avoid downloading it every time

#######################################################
import json
response = requests.get("https://opendomesday.org/api/1.0/county")
if response.status_code == 200:
    counties = response.json()
    for county in counties:
        if county["name"] == "Derbyshire":
            places = county["places_in_county"]
            break
else:
    print("Error occurred while retrieving data from the API.")
    exit()

# Load the details of each place and their manors
for place in places:
    place_id = place["id"]
    place_url = f"https://opendomesday.org/api/1.0/place/{place_id}"
    response_place = requests.get(place_url)
    if response_place.status_code == 200:
        place_details = response_place.json()
        manors = place_details.get("manors", [])

        # Save place details to a file
        with open(f"place_{place_id}.json", "w") as file:
            json.dump(place_details, file)

        print(f"Place ID: {place_id}")
        print("Place Details:")
        print(place_details)
        print("Manor Details:")
        for manor in manors:
            manor_id = manor["id"]
            manor_url = f"https://opendomesday.org/api/1.0/manor/{manor_id}"
            response_manor = requests.get(manor_url)
            if response_manor.status_code == 200:
                manor_details = response_manor.json()

                # Save manor details to a file
                with open(f"manor_{manor_id}.json", "w") as file:
                    json.dump(manor_details, file)

                print(manor_details)
            else:
                print(f"Error occurred while retrieving manor details for ID: {manor_id}")
    else:
        print(f"Error occurred while retrieving place details for ID: {place_id}")
    print("------------------")
#######################################################
```

```python
# Exercise 4
# Now that we have a quantity of raw data, we will extract the interesting parts.  

# - Can you find the corresponding json fields?  
# - Then, you can list these numbers for each manor in Derbyshire.  
# - And format this in an appropriate comma-separated values (CSV) file.

#######################################################

import csv
manor_data = []

# Load the details of each place and their manors
for place in places:
    place_id = place["id"]
    place_url = f"https://opendomesday.org/api/1.0/place/{place_id}"
    response_place = requests.get(place_url)
    if response_place.status_code == 200:
        place_details = response_place.json()
        manors = place_details.get("manors", [])
        for manor in manors:
            manor_id = manor["id"]
            manor_url = f"https://opendomesday.org/api/1.0/manor/{manor_id}"
            response_manor = requests.get(manor_url)
            if response_manor.status_code == 200:
                manor_details = response_manor.json()
                geld = manor_details.get("geld")
                total_ploughs = manor_details.get("totalploughs")
                manor_data.append({"Manor ID": manor_id, "Geld": geld, "Total Ploughs": total_ploughs})
            else:
                print(f"Error occurred while retrieving manor details for ID: {manor_id}")
    else:
        print(f"Error occurred while retrieving place details for ID: {place_id}")

# Write the extracted data to a CSV file
filename = "manor_data.csv"
fields = ["Manor ID", "Geld", "Total Ploughs"]

with open(filename, "w", newline="") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    writer.writerows(manor_data)

print("CSV file created successfully.")
#######################################################

```

```python
# Exercise 8 (optional)
# Add to your class a system for error handling.  
# It must manage the following errors:  
# - Connection error  
# - Parsing error  
# - Request error  
# - Response error  
# - Parameter error

#######################################################
class ManorAnalyzer:
    def __init__(self, filename):
        self.filename = filename
        self.manor_data = []

    def load_data(self):
        try:
            with open(self.filename, "r") as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    self.manor_data.append(row)
        except IOError as e:
            raise ConnectionError("Error loading data from CSV file:", str(e))

    def find_richest_manor(self):
        if not self.manor_data:
            raise ParameterError("No data loaded. Please load data before performing analysis.")

        richest_manor = None
        richest_geld = 0

        for row in self.manor_data:
            geld_str = row["Geld"]
            if geld_str:
                try:
                    geld = float(geld_str)
                    if geld > richest_geld:
                        richest_geld = geld
                        richest_manor = row["Manor ID"]
                except ValueError as e:
                    raise ParsingError("Error parsing 'Geld' value:", str(e))

        if richest_manor is None:
            raise ResponseError("No manor found with a valid 'Geld' value.")

        return richest_manor, richest_geld

    def calculate_total_value(self):
        if not self.manor_data:
            raise ParameterError("No data loaded. Please load data before performing analysis.")

        total_value = 0

        for row in self.manor_data:
            geld_str = row["Geld"]
            if geld_str:
                try:
                    geld = float(geld_str)
                    total_value += geld
                except ValueError as e:
                    raise ParsingError("Error parsing 'Geld' value:", str(e))

        return total_value


filename = "manor_data.csv"
manor_analyzer = ManorAnalyzer(filename)

try:
    manor_analyzer.load_data()
except ConnectionError as e:
    print("Error loading data:", str(e))

try:
    richest_manor, richest_geld = manor_analyzer.find_richest_manor()
    print(f"The richest manor in Derbyshire is Manor ID: {richest_manor} with a geld of {richest_geld}.")
except (ParameterError, ParsingError, ResponseError) as e:
    print("Error finding the richest manor:", str(e))

try:
    total_value = manor_analyzer.calculate_total_value()
    print(f"The total value paid by Derbyshire is: {total_value}.")
except (ParameterError, ParsingError) as e:
    print("Error calculating the total value:", str(e))
#######################################################

```
